```python
import numpy as np
import matplotlib.pylab as plt
%matplotlib inline
import pandas as pd
import sqlite3
```

```python
conn = sqlite3.connect('rotator_1exp_pairsmix_10yrs.db')
```

```python
df = pd.read_sql('select rotTelPos,rotSkyPos from summaryallprops ;', conn)
```

```python
ack = plt.hist(df['rotTelPos'], bins=100)
plt.title('Dithered')
plt.xlabel('rotTelPos (deg)')
plt.savefig('talk_plots/dithered_tel.pdf')
```

```python
ack = plt.hist(df['rotSkyPos'], bins=100)
plt.title('Dithered')
plt.xlabel('rotSkyPos (deg)')
plt.savefig('talk_plots/dithered_sky.pdf')
```

```python
# Let's look at the baseline that had no dithering
conn.close()
conn = sqlite3.connect('../baselines/baseline_1exp_pairsmix_10yrs.db')
df = pd.read_sql('select rotTelPos,rotSkyPos from summaryallprops ;', conn)
```

```python
ack = plt.hist(df['rotTelPos'], bins=100)
plt.title('Not Dithered')
plt.xlabel('rotTelPos (deg)')
plt.savefig('talk_plots/nd_tel.pdf')
```

```python
ack = plt.hist(df['rotSkyPos'], bins=100)
plt.title('Not Dithered')
plt.xlabel('rotSkyPos (deg)')
plt.savefig('talk_plots/nd_sky.pdf')
```
