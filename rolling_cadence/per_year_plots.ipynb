```python
import numpy as np
import healpy as hp
import matplotlib.pylab as plt
%matplotlib inline
import healpy as hp

import lsst.sims.maf.db as db
import lsst.sims.maf.utils as utils
import lsst.sims.maf.metrics as metrics
import lsst.sims.maf.slicers as slicers
import lsst.sims.maf.stackers as stackers
import lsst.sims.maf.metricBundles as metricBundles
```

```python
dbFiles = {'mod2': 'roll_mod2_mixed_10yrs.db', 'mod6': 'roll_mod6_mixed_10yrs.db'}
for dbFile in dbFiles:
    conn = db.OpsimDatabase(dbFiles[dbFile])
    outDir='temp'
    resultsDb = db.ResultsDb(outDir=outDir)


    bundleList = []
    years = [0,1,2,3, 4, 5, 6]
    plotDict = {'percentileClip': 95.}
    for y1,y2 in zip(years[:-1], years[1:]):

        metric = metrics.CountMetric(col='observationStartMJD', metricName='Count')
        slicer = slicers.HealpixSlicer()
        sql = 'night > %i and night < %i' % (365.25*y1, 365.25*y2)
        bundleList.append(metricBundles.MetricBundle(metric,slicer,sql, plotDict=plotDict, runName=dbFile))

    bd = metricBundles.makeBundlesDictFromList(bundleList)
    bg = metricBundles.MetricBundleGroup(bd, conn, outDir=outDir, resultsDb=resultsDb)
    bg.runAll()
    bg.plotAll(closefigs=False)
```
